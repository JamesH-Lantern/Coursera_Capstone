```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
req = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
soup = BeautifulSoup(req.content,'lxml')
table = soup.find_all('table')[0]
df = pd.read_html(str(table))
neighborhood=pd.DataFrame(df[0])
new_nb=neighborhood[neighborhood.Borough != 'Not assigned']
new_nb.head()
```

```python
new_nb.shape
```

```python
#Check to see if any NaN neighborhoods exist with Boroughs that are Not assigned.
test = new_nb[new_nb['Neighborhood'].isna()]
test.shape
```

```python
#In this particular dataset, there are no NaN neighborhoods with assigned Boroughs,
#but if there were, replace NaN with Borough value 
new_nb.fillna(new_nb.Borough, inplace=True)
new_nb
```

```python
#To prove we didnt lose anything
new_nb.shape
```

### Part 2 - Week 3 Peer-Project
Reading in the the Latitude and Longitude file for Toronto Neighborhoods


```python
latlong = pd.read_csv("https://cocl.us/Geospatial_data")
print("Data imported into dataframe: latlong")
```

```python
latlong.head()
```

```python
#Merging the neighborhood dataframe and the Latitude/Longuitude

results = pd.merge(new_nb,
                 latlong[['Postal Code', 'Latitude','Longitude']],
                 on='Postal Code')
results.head()
```

```python
neighborhoods = results[results['Borough'].str.contains("Toronto")]
neighborhoods.head()
```

```python
#Only neighborhoods with 'Toronto' in their Borough name  (i.e. East Toronto, Downtown Toronto, etc)
neighborhoods['Borough'].value_counts()
```

```python
#!conda install -c conda-forge geopy --yes # uncomment this line if you haven't completed the Foursquare API lab
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values

#!conda install -c conda-forge folium=0.5.0 --yes # uncomment this line if you haven't completed the Foursquare API lab
#!pip install folium
import folium # map rendering library

print('Libraries imported.')
```

```python
address = 'Toronto'

geolocator = Nominatim(user_agent="TOR_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinates of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# create map of Toronto using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(neighborhoods['Latitude'], neighborhoods['Longitude'], neighborhoods['Borough'], neighborhoods['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
